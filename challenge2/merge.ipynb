```python
import pandas as pd
import numpy as np
import psycopg2
```

```python
conn = psycopg2.connect("dbname='cpdp' user='cpdp' host='localhost' password='postgres'")
```

```python
officers_df = pd.read_sql('select * from data_officer', con=conn)
officers_df['first_name'] = officers_df['first_name'].str.lower()
officers_df['last_name'] = officers_df['last_name'].str.lower()
officers_df['middle_initial'] = officers_df['middle_initial'].str.lower()
officers_df['suffix_name'] = officers_df['suffix_name'].str.lower()
```

```python
badge_df = pd.read_sql('SELECT * FROM data_officerbadgenumber WHERE current = true', con=conn)
badge_df['star'] = badge_df['star'].astype(float)
```
